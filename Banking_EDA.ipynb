Create Income Category from Estimated Income for plotting

```python
bins = [0 , 100000 , 300000 , float('inf')]
labels = ['Low' , 'Medium', 'High']

df['Income_Category'] = pd.cut(df['Estimated Income'] , bins = bins , labels = labels)
```

```python
df['Income_Category'].value_counts().plot(kind='bar')
```

# **Categorical Columns**

```python
categorical_col = df[['BRId' , 'GenderId', 'IAId' , 'Amount of Credit Cards','Income_Category', 'Nationality' , 'Occupation' , 'Fee Structure' , 'Loyalty Classification' , 'Risk Weighting']]

for cols in categorical_col:
  print(f'Value Counts for {cols}:')
  display(df[cols].value_counts())
  print('\n')
```

# **Numerical Analysis**

```python
numerical_col = df[[ 'Estimated Income','Credit Card Balance', 'Bank Deposits' , 'Superannuation Savings', 'Bank Loans' , 'Checking Accounts' ,'Saving Accounts' , 'Business Lending']]

for cols in numerical_col:
  print(f'Value Counts for {cols}:')
  display(df[cols].value_counts())
  print('\n')
```

# **Univarite Analysis**

```python
plt.figure(figsize=(15, 20))
for i,col in enumerate(numerical_col):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)

plt.show()
```

# **Heatmaps**

```python
numerical_col = ['Estimated Income','Credit Card Balance', 'Bank Deposits',  'Superannuation Savings', 'Bank Loans' , 'Checking Accounts' ,'Saving Accounts' , 'Business Lending'] # Change numerical_col to a list of column names

coorelation_matrix = df[numerical_col].corr() # Now, df[numerical_col] correctly selects the desired columns
plt.figure(figsize=(10, 8))
sns.heatmap(coorelation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

# **Strong Positive Correlations:**

1. Bank Deposits and Checking Accounts (0.84): This means people who have more deposits in their bank accounts tend to have more checking accounts as well. The two are strongly linked.

2. Saving Accounts and Bank Deposits (0.75): People who deposit more money into their bank accounts are likely to also have more savings.


---


**Strategy**

1. Offer more checking accounts to customers who already have a lot of money in their bank. They are more likely to be interested.

2. Promote special deals for customers with big deposits to open extra checking accounts.

3. Encourage customers with big deposits to open savings accounts by offering better interest rates or special deals.
